```python
from IPython.display import HTML

import numpy as np

import matplotlib.pyplot as plt

from matplotlib import rc

from matplotlib.animation import FuncAnimation

import random as rnd

pi = np.pi

rc('animation', html='jshtml')

anims = []

def f1(x, y):
    return 20 + x**2 - 10*np.cos(2*pi*x) + y**2 - 10*np.cos(2*pi*y)
  
def f2(x, y):
    return -20*np.exp(-0.2*(0.5*(x**2+y**2))**0.5) - np.exp(0.5*(np.cos(2*pi*x)+np.cos(2*pi*y))) + np.exp(1) + 20

def f3(x, y):
    return x**2 + y**2

def f4(x, y):
    return 100*(y - x**2)**2 + (x - 1)**2

def f5(x, y):
    return (1.5 - x +x*y)**2 + (2.25 - x + x*y**2)**2 + (2.625 -x + x*y**3)**2


def f6(x, y):
    return (1 + (x+y+1)**2 * (19 -14 *x + 3*x**2 - 14*y +6*x*y + 3*y**2)) * (30 + (2*x -3*y)**2 
                                                                            *(18 - 32*x + 12*x**2 
                                                                              + 48*y -36*x*y + 27*y**2))

def f7(x, y):
    return (x + 2*y - 7)**2 + (2*x + y - 5)**2

def f8(x, y):
    return 100*(abs(y - 0.01*x**2))*0.5 + 0.01*abs(x + 10)

def f9(x, y):
    return 0.26*(x**2 + y**2) - 0.48*x*y

def f10(x, y):
    return np.sin(3*pi*x)**2 + (x-1)**2 * (1 + np.sin(2*pi*y)**2) + (y-1)**2 * (1 + np.sin(2*pi*y)**2)

def f11(x, y):
    return (x**2 + y - 11)**2 + (x + y**2 - 7)**2

def f12(x, y):
    return 2*x**2 -1.05*x**4 + x**6 / 6 + x*y + y**2

def f13(x, y):
    return -np.cos(x)*np.cos(y)*np.exp(-((x - pi)**2 + (y - pi)**2))

def f14(x, y):
    return -0.0001 * (abs(np.sin(x)*np.sin(y)*np.exp(abs(100.0 - 1/pi * (x**2 + y **2)**0.5))) + 1.0)**0.1

def f15(x, y):
    return -(y + 47)*np.sin((abs(x/2 + y + 47))**0.5) - x*np.sin((abs(x - (y+47)))**0.5)

def f16(x, y):
    return -abs(np.sin(x)*np.cos(y)*np.exp(abs(1 - 1/pi * (x**2 + y**2)**0.5)))

def f17(x, y):
    return np.sin(x+y) + (x-y)**2 -1.5*x + 2.5*y + 1

def f18(x, y):
    return 0.5 + (np.sin(x**2 - y**2)**2 - 0.5) / (1 + 0.001*(x**2 + y**2))**2

def f19(x, y):
    return 0.5 + (np.cos(np.sin(abs(x**2 - y**2)))**2 - 0.5) / (1 + 0.001*(x**2 + y**2))**2

def f20(x, y):
    return 0.5*(x**4 - 16*x**2 + 5*x + y**4 - 16*y**2 + 5*y)

def plotter(xbound, ybound, func, xy):
    fig = plt.figure(figsize=(6,5))
    ax = fig.add_subplot(111)
#     ax = fig.add_axes([0, 0, 1, 1], frameon=False)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    
    x = np.linspace(xbound[0], xbound[1], 20)
    y = np.linspace(ybound[0], ybound[1], 20)
    X, Y = np.meshgrid(x, y)
    Z = func(X, Y)  
    
    cont = ax.contourf(X, Y, Z, 20, cmap='RdGy')
    fig.colorbar(cont)
    sc = ax.scatter(xy[0][0][:], xy[0][1][:], marker='o')

    def update(i):
        data = np.c_[xy[i][0][:], xy[i][1][:]]
        sc.set_offsets(data)
    
    anim = FuncAnimation(fig, update, frames=len(xy), interval=10)
    anim
    return anim

def plot_function(xbound, ybound, func):
    #   Функция
    fig = plt.figure(figsize=(10,10))
        
    ax = fig.add_subplot(111, projection='3d')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
                  
    x = np.linspace(xbound[0], xbound[1], 20)
    y = np.linspace(ybound[0], ybound[1], 20)
    X, Y = np.meshgrid(x, y)     
    Z = func(X, Y)
  
    sur = ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                    cmap='viridis', edgecolor='none')

    plt.show()

def plot_conv(solution):
    #   Сходимость
    fig = plt.figure() 
    sol = plt.plot(solution)
    plt.grid(True) 
    plt.xlabel("Era")
    plt.ylabel("Solution")
    plt.show()
    
def swarm_alg(func, xbound, ybound, target, eps, params):
    xmin = xbound[0]
    xmax = xbound[1]
    ymin = ybound[0]
    ymax = ybound[1]
    
    particle_count = params[0]
    max_era = params[1]
    
    solution = list()
    x = list()
    y = list()
    
    location = [[0 for i in range(particle_count)] for j in range(3)]
    velocity = [[0 for i in range(particle_count)] for j in range(2)]
    best_global = [[0 for i in range(particle_count)] for j in range(3)]
    global_min = [[xmax + 1, ymax + 1, 1000000000], [xmax + 1, ymax + 1, 1000000000]]
    init_global = [0, 0, 0]
    
    plotxy = [[[0 for i in range(particle_count)] for j in range(2)] for k in range(max_era)]
            
    p1 = rnd.uniform(0,params[2])
    p2 = rnd.uniform(0,params[3])
    p = p1 + p2
    koef = rnd.uniform(0.000001,0.999999)
    if p > 4:
        hi = 2*koef / abs(2 - (p1 + p2) - ((p1 + p2)**2 - 4 * (p1 + p2))**0.5)
    else:
        hi = koef
        
    yparticle_count = int(((ymax - ymin) * particle_count / (xmax - xmin))**0.5)
    xparticle_count = int((xmax - xmin) * yparticle_count / ( ymax - ymin))
    
    plotflag = params[4]
    
    if plotflag > 0:
        plot_function(xbound, ybound, func)
    
    for i in range(particle_count):
            best_global[0][i] = location[0][i] = rnd.uniform(xmin, xmax)
            best_global[1][i] = location[1][i] = rnd.uniform(ymin, ymax)

            best_global[2][i] = location[2][i] = func(location[0][i], location[1][i])
            
            if best_global[2][i] < global_min[1][2]:
                global_min[1][2] = best_global[2][i]
                global_min[1][0] = best_global[0][i]
                global_min[1][1] = best_global[1][i]
            
            velocity[0][i] = rnd.uniform(-(xmax - xmin), xmax - xmin)
            velocity[1][i] = rnd.uniform(-(ymax - ymin), ymax - ymin)
               
    k = 0
    while k < max_era:
        flag = 0
        init_global[2] = global_min[1][2]
        init_global[0] = global_min[1][0]
        init_global[1] = global_min[1][1]

        for i in range(particle_count):
            p1 = rnd.uniform(0,params[2])
            p2 = rnd.uniform(0,params[3])
            p = p1 + p2
            koef = rnd.uniform(0.000001,0.999999)
            if p > 4:
                hi = 2*koef / abs(2 - (p1 + p2) - ((p1 + p2)**2 - 4 * (p1 + p2))**0.5)
            else:
                hi = koef
    
            velocity[0][i] = hi*(velocity[0][i] + p1 * (best_global[0][i] - location[0][i]) + p2 
                                 * (global_min[1][0] - location[0][i]))
            velocity[1][i] = hi*(velocity[1][i] + p1 * (best_global[1][i] - location[1][i]) + p2
                                 * (global_min[1][1] - location[1][i]))
            location[0][i] += velocity[0][i]
            location[1][i] += velocity[1][i]

            if location[0][i] < xmin:
                temp = xmin - location[0][i] + velocity[0][i]
                location[1][i] -= (location[0][i] - xmin) / velocity[0][i] * velocity[1][i]
                velocity[1][i] -= (location[0][i] - xmin) / velocity[0][i] * velocity[1][i]
                location[0][i] = xmin
                velocity[0][i] = temp 
                
            if location[0][i] > xmax:
                temp = xmax - location[0][i] + velocity[0][i]
                location[1][i] -= (location[0][i] - xmax) / velocity[0][i] * velocity[1][i]
                velocity[1][i] -= (location[0][i] - xmax) / velocity[0][i] * velocity[1][i]
                location[0][i] = xmax
                velocity[0][i] = temp 
                
            if location[1][i] < ymin:
                temp = ymin - location[1][i] + velocity[1][i]
                location[0][i] -= (location[1][i] - ymin) / velocity[1][i] * velocity[0][i]
                velocity[0][i] -= (location[1][i] - ymin) / velocity[1][i] * velocity[0][i]
                location[1][i] = ymin
                velocity[1][i] = temp 
                
            if location[1][i] > ymax:
                temp = ymax - location[1][i] + velocity[1][i]
                location[0][i] -= (location[1][i] - ymax) / velocity[1][i] * velocity[0][i]
                velocity[0][i] -= (location[1][i] - ymax) / velocity[1][i] * velocity[0][i]
                location[1][i] = ymax
                velocity[1][i] = temp 
                
            location[2][i] = func(location[0][i], location[1][i])
            
            if best_global[2][i] > location[2][i]:
                best_global[0][i] = location[0][i]
                best_global[1][i] = location[1][i]
                best_global[2][i] = location[2][i]
                
            if global_min[1][2] > best_global[2][i]:
                flag = 1
                global_min[1][2] = best_global[2][i]
                global_min[1][0] = best_global[0][i]
                global_min[1][1] = best_global[1][i]
            plotxy[k][0][i] = location[0][i]
            plotxy[k][1][i] = location[1][i]
            
        
        if flag == 1:
            global_min[0][0] = init_global[0]
            global_min[0][1] = init_global[1]
            global_min[0][2] = init_global[2]
        
        solution.append(global_min[0][2])
        k += 1
        
    if plotflag > 0:
        anim = plotter(xbound, ybound, func, plotxy)
        
    print("Minimum is ", global_min[0][2])
    flag = 0
    for i in range(particle_count):
        if(abs(global_min[0][2] - target) > eps):
            flag = 1
            break;
    if flag == 0:
        print("Test OK")
    else:
        print("ERROR!!!")
    print("Point is [", global_min[0][0], ",", global_min[0][1], "]")
#     print(p1, p2, hi)
    plot_conv(solution)
    return anim

def start():
    plotflag  =1
    #---------------------------------------------------
    xbound = [-5.12, 5.12]
    ybound = [-5.12, 5.12]
    target = 0.0
    eps = 0.001
    p1 = rnd.uniform(0,5)
    p2 = rnd.uniform(0,5)
    # particles, max_era, p1, p2
    p = [120, 50, p1, p2, plotflag]
    #     p = [100, 100, 2.05, 2.05, 0.7298, plotflag]
    print("1.-------------Функция Растрыгина------------")
    anims.append(swarm_alg(f1, xbound, ybound, target, eps, p))

#     #---------------------------------------------------
#     xbound = [-5, 5]
#     ybound = [-5, 5]
#     target = 0.0
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     p = [100, 100, p1, p2, plotflag]
#     print("2.---------------Функция Экли---------------")
#     swarm_alg(f2, xbound, ybound, target, eps, p)

#     #---------------------------------------------------
#     xbound = [-10000, 10000]
#     ybound = [-10000, 10000]
#     target = 0.0
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     p = [100, 100, p1, p2, plotflag]
#     print("3.---------------Функция сферы---------------")
#     swarm_alg(f3, xbound, ybound, target, eps, p)

#     #---------------------------------------------------
#     xbound = [-10000, 10000]
#     ybound = [-10000, 10000]
#     target = 0.0
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     #     p = [100, 100, p1, p2, plotflag]
#     p = [100, 1000, 2.05, 2.05, plotflag]
#     print("4.------------Функция Розенброка-------------")
#     swarm_alg(f4, xbound, ybound, target, eps, p)

#     #---------------------------------------------------
#     xbound = [-4.5, 4.5]
#     ybound = [-4.5, 4.5]
#     target = 0.0
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     p = [100, 100, p1, p2, plotflag]
#     print("5.---------------Функция Била---------------")
#     swarm_alg(f5, xbound, ybound, target, eps, p)

#     #---------------------------------------------------
#     xbound = [-2, 2]
#     ybound = [-2, 2]
#     target = 3.0
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     p = [100, 100, p1, p2, plotflag]
#     print("6.-----------Функция Гольдман-Прайса--------")
#     swarm_alg(f6, xbound, ybound, target, eps, p)

#     #---------------------------------------------------
#     xbound = [-10, 10]
#     ybound = [-10, 10]
#     target = 0.0
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     p = [100, 100, p1, p2, plotflag]
#     print("7.---------------Функция Бута---------------")
#     swarm_alg(f7, xbound, ybound, target, eps, p)

#     #---------------------------------------------------
#     xbound = [-15, 5]
#     ybound = [-3, 3]
#     target = 0.0
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     #     p = [100, 100, p1, p2, plotflag]
#     p = [100, 1000, 2.05, 2.05, plotflag]
#     print("8.---------------Функция Букина N 6----------")
#     swarm_alg(f8, xbound, ybound, target, eps, p)

#     #---------------------------------------------------
#     xbound = [-10, 10]
#     ybound = [-10, 10]
#     target = 0.0
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     p = [100, 100, p1, p2, plotflag]
#     print("9.---------------Функция Матьяса-------------")
#     swarm_alg(f9, xbound, ybound, target, eps, p)

#     #---------------------------------------------------
#     xbound = [-10, 10]
#     ybound = [-10, 10]
#     target = 0.0
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     p = [100, 100, p1, p2, plotflag]
#     print("10.---------------Функция Леви N 13-----------")
#     swarm_alg(f10, xbound, ybound, target, eps, p)

#     #---------------------------------------------------
#     xbound = [-5, 5]
#     ybound = [-5, 5]
#     target = 0.0
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     p = [100, 100, p1, p2, plotflag]
#     print("11.---------------Функция Химмельблау----------")
#     swarm_alg(f11, xbound, ybound, target, eps, p)

#     #---------------------------------------------------
#     xbound = [-5, 5]
#     ybound = [-5, 5]
#     target = 0.0
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     p = [100, 100, p1, p2, plotflag]
#     print("12.--------Функция трехгорбого верблюда---------")
#     swarm_alg(f12, xbound, ybound, target, eps, p)

#     #---------------------------------------------------
#     xbound = [-100, 100]
#     ybound = [-100, 100]
#     target = -1.0
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     p = [100, 100, p1, p2, plotflag]
#     print("13.---------------Функция Изома-----------------")
#     swarm_alg(f13, xbound, ybound, target, eps, p)

#     #---------------------------------------------------
#     xbound = [-10.0, 10.0]
#     ybound = [-10.0, 10.0]
#     target = -2.06261
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     p = [100, 500, p1, p2, plotflag]
#     print("14.-----------Функция крест на подносе----------")
#     swarm_alg(f14, xbound, ybound, target, eps, p)

#     #---------------------------------------------------
#     xbound = [-512.0, 512.0]
#     ybound = [-512.0, 512.0]
#     target = -959.6407
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     #     p = [100, 100, p1, p2, plotflag]
#     p = [100, 100, 2.05, 2.05, plotflag]
#     print("15.-------Функция Функция подставка для яиц-----")
#     swarm_alg(f15, xbound, ybound, target, eps, p)

#     #---------------------------------------------------
#     xbound = [-10.0, 10.0]
#     ybound = [-10.0, 10.0]
#     target = -19.2085
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     p = [100, 300, p1, p2, plotflag]
#     print("16.-------Табличная функция Хольдера-------------")
#     swarm_alg(f16, xbound, ybound, target, eps, p)

#     #---------------------------------------------------
#     xbound = [-1.5, 4.0]
#     ybound = [-3.0, 4.0]
#     target = -1.9133
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     p = [100, 100, p1, p2, plotflag]
#     print("17.-------Функция МакКормика------------------")
#     swarm_alg(f17, xbound, ybound, target, eps, p)

#     #---------------------------------------------------
#     xbound = [-100.0, 100.0]
#     ybound = [-100.0, 100.0]
#     target = 0.0
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     p = [100, 100, p1, p2, plotflag]
#     print("18.-------Функция Шаффера N2--------------------")
#     swarm_alg(f18, xbound, ybound, target, eps, p)
 
#     #---------------------------------------------------
#     xbound = [-100.0, 100.0]
#     ybound = [-100.0, 100.0]
#     target = 0.292579
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     p = [100, 100, p1, p2, plotflag]
#     print("19.-------Функция Шаффера N4--------------------")
#     swarm_alg(f19, xbound, ybound, target, eps, p)

#     #---------------------------------------------------
#     xbound = [-5.0, 5.0]
#     ybound = [-5.0, 5.0]
#     target = -78.3322
#     eps = 0.001
#     p1 = rnd.uniform(0,5)
#     p2 = rnd.uniform(0,5)
#     # particles, max_era, p1, p2
#     p = [100, 100, p1, p2, plotflag]
#     print("20.-------Функция Стыбинского-Танга-------------")
#     swarm_alg(f20, xbound, ybound, target, eps, p)
    return anims
        
    
anims = start()

for i in range(len(anims)):
    HTML(anims[i].to_jshtml())
```
