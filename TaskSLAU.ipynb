```python
import pandas as pd
import numpy as np
import math
import scipy
from scipy import linalg as lin
import matplotlib
from matplotlib import pyplot

def F(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)

def term(koef, degree, x):
    answer = 0.0
    for i in range(degree + 1):
        answer += koef[i] * x**i
    return answer
```

```python
points = {1,15}
A = np.array([[i**j for j in range(2)] for i in points])
b = np.array([F(i) for i in points])
solution = lin.solve(A, b)

x = np.arange(0.5, 15.5, 0.1)
y = np.array([F(arg) for arg in x])
z = np.array([term(solution, 1, arg) for arg in x])

pyplot.plot(x, y)
pyplot.plot(x, z)
pyplot.show()
```
